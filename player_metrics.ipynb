### Extract players who were traded

```python
from collections import defaultdict

def remove_redundant(data):
    a = defaultdict(list)
    ind_to_remove = []
    for index, row in data.iterrows():
        if row.Player in a[row.Year]:
            ind_to_remove.append(index)
        else:
            a[row.Year].append(row.Player)
    data = data.drop(index=ind_to_remove)
    return data
```

```python
data = remove_redundant(data)
```

On the other hand, we added a few non-advanced player stats inlcuding eFG%. We can see that eFG% is highly correlated with TS%, which can cause problems in our regression model. 

#### TS% vs eFG%:
True shooting factors in free throw attempts and makes, so it is a more reliable stat in determining efficiency as a scorer. If you are mixing in a lot of free throw attempts that will boost your overall TS%. eFG% is mostly about how much you shoot and it applies a weight on threes. If you are a high volume shooter who makes a lot of threes as much as two pointers, your eFG% will be high. For the purpose of evaluating veterans, high volume shooting isn't as much of a concern as efficient scorers, because veterans typically recieve less shot attempts than they would in their "prime" years. We shall remove eFG% from our feature columns.
