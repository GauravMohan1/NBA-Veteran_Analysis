### Extract players who were traded

```python
from collections import defaultdict

def remove_redundant(data):
    a = defaultdict(list)
    ind_to_remove = []
    for index, row in data.iterrows():
        if row.Player in a[row.Year]:
            ind_to_remove.append(index)
        else:
            a[row.Year].append(row.Player)
    data = data.drop(index=ind_to_remove)
    return data
```

```python
data = remove_redundant(data)
```

## Feature Choosing

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

corr = data.corr()
# Make a heatmap with highlights on pairs of features with high correlations.
corr_annot = corr.round(2)
corr_annot[corr_annot.abs()<0.8] = ''
_ = plt.figure(figsize=(16, 12))
_ = sns.heatmap(corr, annot=corr_annot, cmap='coolwarm', fmt='')
```

It appears a lot of these stats are highly correlated with each other. Counting stats are highly correlated to their % counterparts. And because counting stats are highly affected by playing time it only makes sense to use % stats. So let's focus on the advanced stats and look at their heatmaps.

```python
corr = data_2020.corr()
# Make a heatmap with highlights on pairs of features with high correlations.
corr_annot = corr.round(2)
corr_annot[corr_annot.abs()<0.8] = ''
_ = plt.figure(figsize=(16, 12))
_ = sns.heatmap(corr, annot=corr_annot, cmap='coolwarm', fmt='')
```

As you can see. The 2019-2020 season data is advanced stats. DRB % is highly correlated with TRB% so let's just not include total stats for this. Since we have OWS and DWS it doesn't make sense to include WS as that is just a net total of two other features, same goes for BPM. VORP is a highly offensive focused statistic on player creation and scoring per 100 possessions, which is why it is highly correlated with OWS. Let's drop these total metric measures, counting stats that are already accounted for, and advanced stats that are highly correlated.

```python
data = data.drop(['blanl', 'blank2', 'BPM', 'FG', 'FGA', '3P', '3PA', '2P', '2PA', 'FT', 'FTA', 'ORB', 'DRB', 'TRB', 'AST', 'STL',
       'BLK', 'TOV', 'PTS', 'GS', 'VORP', 'TRB%'],axis=1)
```
